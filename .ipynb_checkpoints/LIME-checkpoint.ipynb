```python

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, f1_score, roc_curve, precision_score, recall_score, accuracy_score, roc_auc_score
from sklearn import metrics
from mlxtend.plotting import plot_confusion_matrix
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


names = [ 'CATHARTES AURA', 'COEREBA FLAVEOLA', 'COLUMBA LIVIA', 'CORAGYPS ATRATUS', 'CROTOPHAGA SULCIROSTRIS', 'CYANOCORAX YNCAS', 'EGRETTA THULA', 'FALCO PEREGRINUS', 'FALCO SPARVERIUS', 'HIRUNDO RUSTICA', 'PANDION HALIAETUS', 'PILHERODIUS PILEATUS', 'PITANGUS SULPHURATUS', 'PYRRHOMYIAS CINNAMOMEUS', 'RYNCHOPS NIGER', 'SETOPHAGA FUSCA', 'SYNALLAXIS AZARAE', 'TYRANNUS MELANCHOLICUS']

width_shape=224
height_shape=224
batch_size=16


test_data_dir = 'datasetpreprocesado/test'  

test_datagen = ImageDataGenerator()

test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(width_shape, height_shape), 
    batch_size = batch_size,
    class_mode='categorical', 
    shuffle=False)

custom_Model50= load_model('models/optimizado.keras')

predictions = custom_Model50.predict(test_generator)


y_pred = np.argmax(predictions, axis=1)
y_real = test_generator.classes


matc=confusion_matrix(y_real, y_pred)

plot_confusion_matrix(conf_mat=matc, figsize=(9,9), class_names = names, show_normed=False)
plt.tight_layout()

print(metrics.classification_report(y_real,y_pred, digits = 4))
```

```python
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt

width_shape=224
height_shape=224

names =  [
    'CATHARTES AURA', 'COEREBA FLAVEOLA', 'COLUMBA LIVIA', 'CORAGYPS ATRATUS', 
    'CROTOPHAGA SULCIROSTRIS', 'CYANOCORAX YNCAS', 'EGRETTA THULA', 'FALCO PEREGRINUS',
    'FALCO SPARVERIUS', 'HIRUNDO RUSTICA', 'PANDION HALIAETUS', 'PILHERODIUS PILEATUS',
    'PITANGUS SULPHURATUS', 'PYRRHOMYIAS CINNAMOMEUS', 'RYNCHOPS NIGER', 'SETOPHAGA FUSCA',
    'SYNALLAXIS AZARAE', 'TYRANNUS MELANCHOLICUS']

# Cargar el modelo
modelt = load_model("models\optimizado.keras")
#modelt = custom_vgg_model

# Ruta de la imagen de prueba
imaget_path = 'ImagenPrueba.jpg'

# Leer la imagen, cambiar tamaño y preprocesar
imaget=cv2.resize(cv2.imread(imaget_path), (width_shape, height_shape), interpolation = cv2.INTER_AREA)
xt = np.asarray(imaget)
xt=preprocess_input(xt)
xt = np.expand_dims(xt,axis=0)

# Obtener las predicciones del modelo
preds = modelt.predict(xt)

# Obtener la clase predicha y su porcentaje de confianza
predicted_class_index = np.argmax(preds)
predicted_class_name = names[predicted_class_index]
confidence_percentage = preds[0][predicted_class_index] * 100

# Imprimir el resultado
print(f'Clase predicha: {predicted_class_name}')
print(f'Porcentaje de confianza: {confidence_percentage:.2f}%')

# Mostrar la imagen
plt.imshow(cv2.cvtColor(np.asarray(imaget), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import tensorflow as tf
from tensorflow.keras.models import Model

# Cargar el modelo pre-entrenado (puedes cambiar esto por tu propio modelo)
model = load_model("models\optimizado.keras")



# Obtener la última capa convolucional
last_conv_layer = model.get_layer('block5_conv3')

# Crear un modelo que devuelva tanto las salidas de la última capa convolucional como las predicciones del modelo
grad_model = Model(inputs=model.inputs, outputs=[last_conv_layer.output, model.output])


def get_gradcam(image_path):
    # Cargar la imagen
    img = image.load_img(image_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)

    # Obtener las predicciones del modelo
    preds = model.predict(img)
    class_idx = np.argmax(preds[0])

    # Obtener las salidas de la última capa convolucional y las predicciones del modelo
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img)
        loss = predictions[:, class_idx]

    # Calcular gradientes
    grads = tape.gradient(loss, conv_outputs)[0]

    # Obtener los pesos de los gradientes
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]

    # Multiplicar cada mapa de características por su peso de gradiente y agregarlos
    for i in range(conv_outputs.shape[-1]):
        conv_outputs[:, :, i] *= pooled_grads[i]

    # Obtener el mapa de calor
    heatmap = np.mean(conv_outputs, axis=-1)
    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap)

    # Resize del mapa de calor para que coincida con el tamaño de la imagen original
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[2]))

    # Convertir el mapa de calor a imagen de color
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # Superponer el mapa de calor en la imagen original
    superimposed_img = cv2.addWeighted(img[0], 0.5, heatmap, 0.5, 0)

    return img[0], superimposed_img

# Ruta de la imagen de prueba
image_path = 'ImagenPrueba.jpg'

# Generar Grad-CAM para la imagen de prueba
original_img, gradcam_img = get_gradcam(image_path)

# Visualizar la imagen original y el Grad-CAM
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradcam_img)
plt.title('Grad-CAM')
plt.axis('off')

plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
from sklearn.metrics import confusion_matrix

# Cargar el modelo
model_path = "models/optimizado.keras"
model = tf.keras.models.load_model(model_path)

# Directorio donde están las imágenes de prueba
test_dir = "datasetpreprocesado/valid"

# Obtener todas las clases (nombres de las carpetas)
class_names = sorted(os.listdir(test_dir))

# Preparar los datos de prueba
X_test = []
y_test = []
for i, class_name in enumerate(class_names):
    class_dir = os.path.join(test_dir, class_name)
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        X_test.append(img_array)
        y_test.append(i)

X_test = np.array(X_test)
y_test = np.array(y_test)

# Hacer predicciones en el conjunto de prueba
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Calcular sensibilidad para cada clase
sensitivity = {}
for i in range(len(conf_matrix)):
    tp = conf_matrix[i, i]
    fn = np.sum(conf_matrix[i, :]) - tp
    sensitivity[i] = tp / (tp + fn)

# Imprimir sensibilidad para cada clase
for i, class_name in enumerate(class_names):
    print(f"Sensibilidad para la clase {class_name}: {sensitivity[i]:.4f}")
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2

# Cargar el modelo entrenado
model_path = "models/optimizado.keras"
model = tf.keras.models.load_model(model_path)

# Directorio donde se encuentran las imágenes de prueba
test_dir = "datasetpreprocesado/test"

# Obtener los nombres de las clases (nombres de las carpetas)
class_names =[ 'CATHARTES AURA', 'COEREBA FLAVEOLA', 'COLUMBA LIVIA', 'CORAGYPS ATRATUS', 'CROTOPHAGA SULCIROSTRIS', 
         'CYANOCORAX YNCAS', 'EGRETTA THULA', 'FALCO PEREGRINUS', 'FALCO SPARVERIUS', 'HIRUNDO RUSTICA', 'PANDION HALIAETUS',
         'PILHERODIUS PILEATUS', 'PITANGUS SULPHURATUS', 'PYRRHOMYIAS CINNAMOMEUS', 'RYNCHOPS NIGER', 
         'SETOPHAGA FUSCA', 'SYNALLAXIS AZARAE', 'TYRANNUS MELANCHOLICUS'] 

# Cargar y redimensionar las imágenes de prueba
X_test_resized = []
for class_name in class_names:
    class_dir = os.path.join(test_dir, class_name)
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        X_test_resized.append(img_array)

X_test_resized = np.array(X_test_resized)

# Obtener la última capa convolucional
last_conv_layer = model.get_layer("block5_conv3")

# Calcular los gradientes
with tf.GradientTape() as tape:
    last_conv_output = last_conv_layer(X_test_resized)
    tape.watch(last_conv_output)
    preds = model(X_test_resized)
    class_idx = np.argmax(preds[0])  # Índice de la clase objetivo

grads = tape.gradient(preds[:, class_idx], last_conv_output)
pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

# Calcular el mapa de calor
heatmap = tf.reduce_mean(last_conv_output, axis=-1)
heatmap = tf.maximum(heatmap, 0)
heatmap /= tf.reduce_max(heatmap)
heatmap *= pooled_grads

# Visualizar el mapa de calor
plt.imshow(heatmap[0])
plt.title(f"Heatmap para la clase {class_names[class_idx]}")
plt.show()

# Superponer el mapa de calor en la imagen original
for i in range(len(X_test_resized)):
    img = X_test_resized[i]
    heatmap_resized = cv2.resize(heatmap[i], (img.shape[1], img.shape[0]))
    heatmap_colormap = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(img, 0.6, heatmap_colormap, 0.4, 0)

    plt.imshow(superimposed_img)
    plt.title(f"Superpuesto para la clase {class_names[class_idx]}")
    plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2

# Cargar el modelo entrenado
model_path = "models/optimizado.keras"
model = tf.keras.models.load_model(model_path)

# Directorio donde se encuentran las imágenes de prueba
test_dir = "datasetpreprocesado/test"

# Obtener los nombres de las clases (nombres de las carpetas)
class_names = sorted(os.listdir(test_dir))

# Cargar y redimensionar las imágenes de prueba en lotes más pequeños
batch_size = 16
X_test_resized = []

for class_name in class_names:
    class_dir = os.path.join(test_dir, class_name)
    img_paths = [os.path.join(class_dir, img_name) for img_name in os.listdir(class_dir)]
    
    for i in range(0, len(img_paths), batch_size):
        batch_img_paths = img_paths[i : i + batch_size]
        batch_imgs = [tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224)) for img_path in batch_img_paths]
        batch_img_arrays = [tf.keras.preprocessing.image.img_to_array(img) for img in batch_imgs]
        X_test_resized.extend(batch_img_arrays)

X_test_resized = np.array(X_test_resized)

# Obtener la última capa convolucional
last_conv_layer = model.get_layer("block5_conv3")

# Calcular los gradientes
with tf.GradientTape() as tape:
    last_conv_output = last_conv_layer(X_test_resized)
    tape.watch(last_conv_output)
    preds = model(X_test_resized)
    class_idx = np.argmax(preds[0])  # Índice de la clase objetivo

grads = tape.gradient(preds[:, class_idx], last_conv_output)
pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

# Calcular el mapa de calor
heatmap = tf.reduce_mean(last_conv_output, axis=-1)
heatmap = tf.maximum(heatmap, 0)
heatmap /= tf.reduce_max(heatmap)
heatmap *= pooled_grads

# Visualizar el mapa de calor
plt.imshow(heatmap[0])
plt.title(f"Heatmap para la clase {class_names[class_idx]}")
plt.show()

# Superponer el mapa de calor en la imagen original
for i in range(len(X_test_resized)):
    img = X_test_resized[i]
    heatmap_resized = cv2.resize(heatmap[i], (img.shape[1], img.shape[0]))
    heatmap_colormap = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(img, 0.6, heatmap_colormap, 0.4, 0)

    plt.imshow(superimposed_img)
    plt.title(f"Superpuesto para la clase {class_names[class_idx]}")
    plt.show()
```
