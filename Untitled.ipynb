```python
import os
import matplotlib.pyplot as plt
from lime import lime_image
from skimage.io import imread
from skimage.transform import resize
from skimage.segmentation import mark_boundaries
import numpy as np

# Función para cargar y preprocesar una imagen
def load_and_preprocess_image(image_path, target_size):
    # Cargar la imagen
    img = imread(image_path)
    
    # Preprocesar la imagen (por ejemplo, redimensionar)
    img = resize(img, target_size)
    
    return img

# Ruta base donde se encuentran las carpetas de cada clase
base_path = 'datasetpreprocesado/test'

# Lista de carpetas/clases dentro de la ruta base
clases = os.listdir(base_path)

# Configurar LimeImageExplainer
explainer = lime_image.LimeImageExplainer()

# Definir el tamaño deseado para las imágenes (ejemplo: 224x224)
target_size = (224, 224)

# Configurar la visualización de las imágenes y explicaciones
num_cols = 6  # Mostrar las explicaciones para las 6 clases principales
fig, axs = plt.subplots(len(clases), num_cols, figsize=(15, len(clases) * 3))
axs = axs.flatten()  # Aplanar la matriz de ejes para iterar fácilmente

# Colores para cada clase (puedes definir más colores según tus necesidades)
class_colors = ['red', 'blue', 'green', 'orange', 'purple', 'yellow']

# Diccionario para mapear nombres de colores a valores RGB numéricos
color_map = {
    'red': (1, 0, 0),
    'blue': (0, 0, 1),
    'green': (0, 1, 0),
    'orange': (1, 0.5, 0),
    'purple': (0.5, 0, 0.5),
    'yellow': (1, 1, 0)
}

# Iterar sobre las carpetas/clases
for i, clase in enumerate(clases):
    # Ruta completa de la carpeta/clase actual
    carpeta_clase = os.path.join(base_path, clase)
    
    # Lista de nombres de las imágenes en la carpeta/clase actual
    imagenes = os.listdir(carpeta_clase)
    
    # Iterar sobre las imágenes en la carpeta/clase actual y mostrarlas con sus explicaciones
    for j, imagen in enumerate(imagenes[:num_cols]):
        ruta_imagen = os.path.join(carpeta_clase, imagen)
        
        # Preprocesar la imagen
        imagen_prueba = load_and_preprocess_image(ruta_imagen, target_size)
        
        # Explicar una predicción individual
        explanation = explainer.explain_instance(
            imagen_prueba,
            modelo.predict,  # La función de predicción de tu modelo
            top_labels=6,  # El número de etiquetas principales a explicar
            hide_color=0,  # El color para ocultar partes de la imagen
            num_samples=1000  # Número de muestras de permutaciones
        )

        # Obtener las imágenes explicadas para cada clase
        for k, label in enumerate(explanation.top_labels):
            temp, mask = explanation.get_image_and_mask(
                label,
                positive_only=True,
                num_features=20,  # Ajustar el número de características
                hide_rest=False
            )

            # Mostrar la imagen con las explicaciones superpuestas
            axs[i*num_cols + j].imshow(mark_boundaries(temp / 255.0, mask, outline_color=color_map[class_colors[k]]))
            axs[i*num_cols + j].axis('off')
            axs[i*num_cols + j].set_title(f"Clase {label}")

# Deshabilitar los ejes para cualquier subtrama adicional que no esté llena
for ax in axs[len(clases)*num_cols:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import subprocess

# Obtener el nombre del entorno actual
env_name = "Python 3"  # Reemplaza con el nombre de tu entorno

# Ejecutar el comando para obtener la lista de paquetes instalados en el entorno
command = f"conda list -n {env_name} --export"

# Ejecutar el comando y capturar la salida
process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
output, error = process.communicate()

# Verificar errores en la ejecución del comando
if error:
    print(f"Error al obtener las librerías del entorno {env_name}: {error.decode()}")
else:
    # Decodificar la salida y dividirla en líneas
    output_str = output.decode()
    packages = output_str.strip().split("\n")

    # Filtrar solo las librerías que has instalado explícitamente (no las por defecto)
    important_packages = [pkg for pkg in packages if not pkg.startswith("#") and not pkg.startswith("_")]

    # Guardar las librerías importantes en un archivo requirements.txt
    with open("requerimientosv2.txt", "w") as f:
        for pkg in important_packages:
            f.write(pkg + "\n")

    print(f"Librerías importantes del entorno {env_name} guardadas en requerimientos.txt")

```

```python
import os
import subprocess

# Obtener el nombre del entorno actual de Conda
process = subprocess.Popen('conda env list', stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
output, error = process.communicate()

if error:
    print(f"Error al obtener el listado de entornos: {error.decode()}")
else:
    env_lines = output.decode().strip().split('\n')
    current_env = None

    for line in env_lines:
        if '*' in line:
            current_env = line.split()[0]
            break

    if current_env is None:
        print("No se pudo detectar el entorno actual de Conda.")
    else:
        # Obtener la lista de paquetes instalados en el entorno actual
        command = f"conda list -n {current_env} --export"
        process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
        output, error = process.communicate()

        if error:
            print(f"Error al obtener las librerías del entorno {current_env}: {error.decode()}")
        else:
            # Decodificar la salida y dividirla en líneas
            output_str = output.decode()
            packages = output_str.strip().split("\n")

            # Filtrar solo las librerías que has instalado explícitamente (no las por defecto)
            important_packages = [pkg for pkg in packages if not pkg.startswith("#") and not pkg.startswith("_")]

            # Guardar las librerías importantes en un archivo requirements.txt
            with open("requerimientosv.txt", "w") as f:
                for pkg in important_packages:
                    f.write(pkg + "\n")

            print(f"Librerías importantes del entorno {current_env} guardadas en requerimientos.txt")
```
