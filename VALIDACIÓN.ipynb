# preprocesar una imagen  a 224x224

```python
from PIL import Image

def preprocess_image(image_path, target_size=(224, 224)):
    # Cargar la imagen
    image = Image.open(image_path)
    
    # Obtener las dimensiones de la imagen
    original_width, original_height = image.size
    
    # Paso 1: Reducir el tamaño de la imagen si es muy grande
    max_dimension = max(original_width, original_height)
    if max_dimension > 1024:  # Puedes ajustar este valor según tus necesidades
        reduction_ratio = 1024 / max_dimension
        new_size = (int(original_width * reduction_ratio), int(original_height * reduction_ratio))
        image = image.resize(new_size, Image.LANCZOS)
    
    # Obtener las nuevas dimensiones de la imagen
    new_width, new_height = image.size
    
    # Paso 2: Recortar la imagen para ajustarla a 224x224 píxeles
    aspect_ratio = new_width / new_height
    target_width, target_height = target_size
    
    if aspect_ratio > 1:  # Imagen horizontal
        new_width = int(aspect_ratio * target_height)
        image = image.resize((new_width, target_height), Image.LANCZOS)
        left = (new_width - target_width) / 2
        top = 0
        right = (new_width + target_width) / 2
        bottom = target_height
    else:  # Imagen vertical o cuadrada
        new_height = int(target_width / aspect_ratio)
        image = image.resize((target_width, new_height), Image.LANCZOS)
        left = 0
        top = (new_height - target_height) / 2
        right = target_width
        bottom = (new_height + target_height) / 2
    
    # Recortar la imagen
    image = image.crop((left, top, right, bottom))
    
    # Guardar o retornar la imagen preprocesada
    image.save('preprocessed_image.jpg')  # Puedes ajustar el nombre del archivo de salida según tus necesidades
    return image

# Ejemplo de uso
preprocessed_image = preprocess_image('datasetpreprocesado/test3/CATHARTES AURA/CATHARTES_AURA_6.jpg')
preprocessed_image.show()  # Mostrar la imagen preprocesada
```

# preprocesar imagenes masivamente a 224x224

```python
import os
from PIL import Image

def preprocess_image(image_path, target_size=(224, 224)):
    # Cargar la imagen
    image = Image.open(image_path)
    
    # Obtener las dimensiones de la imagen
    original_width, original_height = image.size
    
    # Paso 1: Reducir el tamaño de la imagen si es muy grande
    max_dimension = max(original_width, original_height)
    if max_dimension > 1024:  # Puedes ajustar este valor según tus necesidades
        reduction_ratio = 1024 / max_dimension
        new_size = (int(original_width * reduction_ratio), int(original_height * reduction_ratio))
        image = image.resize(new_size, Image.LANCZOS)
    
    # Obtener las nuevas dimensiones de la imagen
    new_width, new_height = image.size
    
    # Paso 2: Recortar la imagen para ajustarla a 224x224 píxeles
    aspect_ratio = new_width / new_height
    target_width, target_height = target_size
    
    if aspect_ratio > 1:  # Imagen horizontal
        new_width = int(aspect_ratio * target_height)
        image = image.resize((new_width, target_height), Image.LANCZOS)
        left = (new_width - target_width) / 2
        top = 0
        right = (new_width + target_width) / 2
        bottom = target_height
    else:  # Imagen vertical o cuadrada
        new_height = int(target_width / aspect_ratio)
        image = image.resize((target_width, new_height), Image.LANCZOS)
        left = 0
        top = (new_height - target_height) / 2
        right = target_width
        bottom = (new_height + target_height) / 2
    
    # Recortar la imagen
    image = image.crop((left, top, right, bottom))
    
    return image

def process_directory(directory_path, output_directory, target_size=(224, 224)):
    # Crear el directorio de salida si no existe
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif')):
                file_path = os.path.join(root, file)
                relative_path = os.path.relpath(file_path, directory_path)
                output_path = os.path.join(output_directory, relative_path)

                # Crear el directorio de salida si no existe
                output_dir = os.path.dirname(output_path)
                if not os.path.exists(output_dir):
                    os.makedirs(output_dir)

                # Procesar y guardar la imagen
                try:
                    processed_image = preprocess_image(file_path, target_size)
                    processed_image.save(output_path)
                    print(f"Processed and saved: {output_path}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Ejemplo de uso
input_directory = r'C:\Users\DIAZOVIEDO\Documents\GitHub\Clasificacion_Alada\datasetpreprocesado\train2'
output_directory = r'C:\Users\DIAZOVIEDO\Documents\GitHub\Clasificacion_Alada\datasetpreprocesado\train2r'
process_directory(input_directory, output_directory)
```

```python
import cv2
import numpy as np
import os

# Función para desenfocar el fondo de una imagen
def desenfocar_fondo(imagen):
    # Verificar si la imagen fue cargada correctamente
    if imagen is None:
        print("Error: la imagen no se ha cargado correctamente.")
        return None

    # Convertir la imagen al espacio de color HSV
    hsv = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)
    
    # Definir el rango de color para el ave (ajustar estos valores según sea necesario)
    lower_color = np.array([0, 50, 50])
    upper_color = np.array([30, 255, 255])
    
    # Crear una máscara para los colores del ave
    mascara = cv2.inRange(hsv, lower_color, upper_color)
    
    # Invertir la máscara para obtener el fondo
    mascara_invertida = cv2.bitwise_not(mascara)
    
    # Mejorar la máscara utilizando la detección de contornos
    contornos, _ = cv2.findContours(mascara, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    mascara_refinada = np.zeros_like(mascara)
    cv2.drawContours(mascara_refinada, contornos, -1, (255), thickness=cv2.FILLED)
    
    # Invertir la máscara refinada
    mascara_refinada_invertida = cv2.bitwise_not(mascara_refinada)
    
    # Aplicar un desenfoque gaussiano a la imagen original
    desenfoque = cv2.GaussianBlur(imagen, (21, 21), 0)
    
    # Mantener la parte de la imagen original que no está desenfocada (el ave)
    imagen_sin_desenfoque = cv2.bitwise_and(imagen, imagen, mask=mascara_refinada)
    
    # Combinar la parte desenfocada con la parte no desenfocada
    imagen_con_desenfoque = cv2.add(imagen_sin_desenfoque, cv2.bitwise_and(desenfoque, desenfoque, mask=mascara_refinada_invertida))
    
    return imagen_con_desenfoque

# Directorio raíz
directorio_raiz = 'datasetpreprocesado/train2'

# Directorio de salida
directorio_salida = 'datasetpreprocesado/train2r_desenfocado'

# Crear el directorio de salida si no existe
os.makedirs(directorio_salida, exist_ok=True)

# Recorrer todos los subdirectorios y archivos
for subdir, _, archivos in os.walk(directorio_raiz):
    for archivo in archivos:
        if archivo.endswith(('.jpg', '.jpeg', '.png')):  # Procesar solo imágenes
            ruta_imagen = os.path.join(subdir, archivo)
            
            # Leer la imagen
            imagen = cv2.imread(ruta_imagen)
            
            # Verificar si la imagen se ha cargado correctamente
            if imagen is None:
                print(f"Error al cargar la imagen: {ruta_imagen}")
                continue
            
            # Desenfocar el fondo
            imagen_desenfocada = desenfocar_fondo(imagen)
            
            # Verificar si se desenfocó correctamente
            if imagen_desenfocada is None:
                print(f"Error al procesar la imagen: {ruta_imagen}")
                continue
            
            # Crear el nuevo path manteniendo la estructura de directorios
            subdir_relativo = os.path.relpath(subdir, directorio_raiz)
            nuevo_subdir = os.path.join(directorio_salida, subdir_relativo)
            os.makedirs(nuevo_subdir, exist_ok=True)
            
            # Guardar la imagen desenfocada en el nuevo directorio
            ruta_nueva_imagen = os.path.join(nuevo_subdir, archivo)
            cv2.imwrite(ruta_nueva_imagen, imagen_desenfocada)

print("Proceso de desenfoque completado.")
```
